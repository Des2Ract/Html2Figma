```python
import os
import gc
import psutil
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import numpy as np
import pandas as pd
import h5py
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
from tqdm import tqdm
import ast
import warnings
from torch.cuda.amp import autocast, GradScaler
import pickle
import json
from collections import defaultdict
import time

warnings.filterwarnings('ignore')

# Memory monitoring function
def print_memory_usage(printing=True):
    process = psutil.Process(os.getpid())
    memory_info = process.memory_info()
    memory_gb = memory_info.rss / 1024**3
    if printing:
        print(f"Memory usage: {memory_gb:.2f} GB")
    return memory_gb

# Set memory limit and device
MEMORY_LIMIT_GB = 7.5
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Initialize GradScaler for mixed precision
scaler = GradScaler()

class EarlyStopping:
    """Early stopping to prevent overfitting."""
    def __init__(self, patience=7, min_delta=0.0001, restore_best_weights=True):
        self.patience = patience
        self.min_delta = min_delta
        self.restore_best_weights = restore_best_weights
        self.best_loss = None
        self.counter = 0
        self.best_weights = None
        
    def __call__(self, val_loss, model):
        if self.best_loss is None:
            self.best_loss = val_loss
            self.save_checkpoint(model)
        elif val_loss < self.best_loss - self.min_delta:
            self.best_loss = val_loss
            self.counter = 0
            self.save_checkpoint(model)
        else:
            self.counter += 1
            
        if self.counter >= self.patience:
            if self.restore_best_weights:
                model.load_state_dict(self.best_weights)
            return True
        return False
    
    def save_checkpoint(self, model):
        """Save model checkpoint."""
        self.best_weights = model.state_dict().copy()

class LazyFigmaDataset(Dataset):
    """Lazy loading dataset for Figma node sequences."""
    
    def __init__(self, data_path, sequence_metadata, label_encoder, cache_size=100):
        """
        Args:
            data_path: Path to the data file
            sequence_metadata: List of dicts with sequence info (start_idx, end_idx, seq_id)
            label_encoder: Fitted label encoder
            cache_size: Number of sequences to keep in memory cache
        """
        self.data_path = data_path
        self.sequence_metadata = sequence_metadata
        self.label_encoder = label_encoder
        self.cache_size = cache_size
        self.cache = {}
        self.cache_order = []
        
        # Store file handle for efficient access
        self._file_handle = None
        self._open_file()
    
    def _open_file(self):
        """Open file handle based on file type."""
        if self.data_path.endswith('.csv'):
            # For CSV, we'll read in chunks as needed
            self._file_handle = None
        elif self.data_path.endswith(('.hdf5', '.h5')):
            self._file_handle = h5py.File(self.data_path, 'r')
    
    def _close_file(self):
        """Close file handle."""
        if self._file_handle is not None:
            self._file_handle.close()
            self._file_handle = None
    
    def __del__(self):
        """Cleanup file handle."""
        self._close_file()
    
    def _load_sequence_from_file(self, seq_metadata):
        """Load a single sequence from file."""
        start_idx = seq_metadata['start_idx']
        end_idx = seq_metadata['end_idx']
        
        if self.data_path.endswith('.csv'):
            # Read specific rows from CSV
            df_chunk = pd.read_csv(
                self.data_path, 
                skiprows=range(1, start_idx + 1),  # Skip header + rows before start
                nrows=end_idx - start_idx,
                header=0 if start_idx == 0 else None
            )
            if start_idx > 0:
                # Set proper column names if we skipped the header
                df_full = pd.read_csv(self.data_path, nrows=1)
                df_chunk.columns = df_full.columns
            
            df_chunk['feature_vector'] = df_chunk['feature_vector'].apply(ast.literal_eval).apply(np.array)
            feature_vectors = np.stack(df_chunk['feature_vector'].values)
            tags = df_chunk['tag'].values
            
        elif self.data_path.endswith(('.hdf5', '.h5')):
            # Read from HDF5
            if self._file_handle is None:
                self._open_file()
            
            feature_vectors = self._file_handle['feature_vector'][start_idx:end_idx]
            tags = [s.decode('utf-8') for s in self._file_handle['tag'][start_idx:end_idx]]
        
        # Convert to tensors
        labels = self.label_encoder.transform(tags)
        
        return {
            'features': torch.FloatTensor(feature_vectors),
            'labels': torch.LongTensor(labels)
        }
    
    def _manage_cache(self, seq_id, sequence_data):
        """Manage LRU cache for sequences."""
        # Remove from cache if already exists (to update order)
        if seq_id in self.cache:
            self.cache_order.remove(seq_id)
        
        # Add to cache
        self.cache[seq_id] = sequence_data
        self.cache_order.append(seq_id)
        
        # Remove oldest if cache is full
        while len(self.cache) > self.cache_size:
            oldest_seq_id = self.cache_order.pop(0)
            del self.cache[oldest_seq_id]
    
    def __len__(self):
        return len(self.sequence_metadata)
    
    def __getitem__(self, idx):
        seq_metadata = self.sequence_metadata[idx]
        seq_id = seq_metadata['seq_id']
        
        # Check cache first
        if seq_id in self.cache:
            # Move to end of cache order (most recently used)
            self.cache_order.remove(seq_id)
            self.cache_order.append(seq_id)
            sequence_data = self.cache[seq_id]
        else:
            # Load from file
            sequence_data = self._load_sequence_from_file(seq_metadata)
            self._manage_cache(seq_id, sequence_data)
        
        return {
            'features': sequence_data['features'],
            'labels': sequence_data['labels'],
            'seq_id': seq_id
        }

def create_sequence_metadata(data_path, max_sequences=None):
    """Create metadata for sequences without loading all data into memory."""
    print(f"Creating sequence metadata from {data_path}...")
    
    sequence_metadata = []
    all_tags = set()
    
    if data_path.endswith('.csv'):
        # Read CSV in chunks to find sequence boundaries
        chunk_size = 10000
        current_idx = 0
        sequence_start = 0
        
        for chunk in pd.read_csv(data_path, chunksize=chunk_size):
            for i, row in chunk.iterrows():
                all_tags.add(row['tag'])
                
                if row['tag'] == 'E_WEBSITE':
                    # End of sequence
                    sequence_metadata.append({
                        'seq_id': f'seq_{len(sequence_metadata)}',
                        'start_idx': sequence_start,
                        'end_idx': current_idx + 1,
                        'length': current_idx + 1 - sequence_start
                    })
                    sequence_start = current_idx + 1
                    
                    if max_sequences and len(sequence_metadata) >= max_sequences:
                        break
                
                current_idx += 1
            
            if max_sequences and len(sequence_metadata) >= max_sequences:
                break
    
    elif data_path.endswith(('.hdf5', '.h5')):
        # Read HDF5 metadata
        with h5py.File(data_path, 'r') as f:
            tags = f['tag'][:]
            sequence_start = 0
            
            for i, tag_bytes in enumerate(tags):
                tag = tag_bytes.decode('utf-8')
                all_tags.add(tag)
                
                if tag == 'E_WEBSITE':
                    sequence_metadata.append({
                        'seq_id': f'seq_{len(sequence_metadata)}',
                        'start_idx': sequence_start,
                        'end_idx': i + 1,
                        'length': i + 1 - sequence_start
                    })
                    sequence_start = i + 1
                    
                    if max_sequences and len(sequence_metadata) >= max_sequences:
                        break
    
    print(f"Found {len(sequence_metadata)} sequences")
    print(f"Found {len(all_tags)} unique tags")
    
    # Analyze sequence lengths
    lengths = [seq['length'] for seq in sequence_metadata]
    print(f"Sequence length stats:")
    print(f"  Min: {min(lengths)}, Max: {max(lengths)}")
    print(f"  Mean: {np.mean(lengths):.2f}, Median: {np.median(lengths):.2f}")
    print(f"  95th percentile: {np.percentile(lengths, 95):.2f}")
    
    return sequence_metadata, sorted(list(all_tags))

def lazy_collate_fn(batch):
    """Memory-efficient collate function for lazy loading."""
    features = [item['features'] for item in batch]
    labels = [item['labels'] for item in batch]
    seq_ids = [item['seq_id'] for item in batch]
    
    lengths = torch.tensor([len(f) for f in features])
    max_len = max(lengths)
    
    feature_dim = features[0].shape[1]
    
    # Create padded tensors
    padded_features = torch.zeros((len(batch), max_len, feature_dim), dtype=torch.float32)
    padded_labels = torch.full((len(batch), max_len), -100, dtype=torch.long)
    
    for i, (f, l) in enumerate(zip(features, labels)):
        seq_len = min(f.shape[0], max_len)
        padded_features[i, :seq_len] = f[:seq_len]
        padded_labels[i, :seq_len] = l[:seq_len]
        lengths[i] = seq_len
    
    return {
        'features': padded_features,
        'labels': padded_labels,
        'lengths': lengths,
        'seq_ids': seq_ids
    }

class OptimizedFigmaBLSTM(nn.Module):
    """Memory-optimized Bidirectional LSTM model."""
    
    def __init__(self, input_dim, hidden_dim, output_dim, num_layers=1, dropout=0.3):
        super(OptimizedFigmaBLSTM, self).__init__()
        
        self.hidden_dim = hidden_dim
        self.num_layers = num_layers
        
        self.lstm = nn.LSTM(
            input_dim, 
            hidden_dim, 
            num_layers=num_layers,
            batch_first=True, 
            bidirectional=True,
            dropout=dropout if num_layers > 1 else 0
        )
        
        self.batch_norm = nn.BatchNorm1d(hidden_dim * 2)
        self.fc = nn.Linear(hidden_dim * 2, output_dim)
        self.dropout = nn.Dropout(dropout)
    
    def forward(self, x, lengths):
        batch_size, seq_len, _ = x.size()
        
        # Pack padded sequence
        packed_input = nn.utils.rnn.pack_padded_sequence(
            x, lengths.cpu(), batch_first=True, enforce_sorted=False
        )
        
        # LSTM forward pass
        packed_output, _ = self.lstm(packed_input)
        output, _ = nn.utils.rnn.pad_packed_sequence(packed_output, batch_first=True)
        
        # Apply dropout
        output = self.dropout(output)
        
        # Batch normalization (if batch size > 1)
        if output.size(0) > 1:
            output_reshaped = output.reshape(-1, output.size(-1))
            output_reshaped = self.batch_norm(output_reshaped)
            output = output_reshaped.reshape(batch_size, seq_len, -1)
        
        # Final classification layer
        logits = self.fc(output)
        
        return logits

def safe_training_step(model, batch, criterion, optimizer, device):
    """Safe training step with mixed precision and memory management."""
    try:
        features = batch['features'].to(device)
        labels = batch['labels'].to(device)
        lengths = batch['lengths']
        
        optimizer.zero_grad()
        
        with autocast():
            outputs = model(features, lengths)
            batch_size, seq_len, num_classes = outputs.size()
            outputs = outputs.reshape(-1, num_classes)
            labels = labels.reshape(-1)
            loss = criterion(outputs, labels)
        
        scaler.scale(loss).backward()
        scaler.unscale_(optimizer)
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        scaler.step(optimizer)
        scaler.update()
        
        # Clean up GPU memory
        del features, labels, outputs
        if torch.cuda.is_available():
            torch.cuda.empty_cache()
        
        return loss.item()
    
    except RuntimeError as e:
        if "out of memory" in str(e):
            print(f"WARNING: Out of memory error occurred: {e}")
            if torch.cuda.is_available():
                torch.cuda.empty_cache()
            return None
        else:
            raise e

def evaluate_model(model, data_loader, criterion, device, label_encoder, phase="Validation"):
    """Comprehensive model evaluation with metrics."""
    model.eval()
    total_loss = 0.0
    all_predictions = []
    all_labels = []
    all_probabilities = []
    successful_batches = 0
    
    with torch.no_grad():
        progress_bar = tqdm(data_loader, desc=f"{phase} Evaluation")
        for batch in progress_bar:
            try:
                features = batch['features'].to(device)
                labels = batch['labels'].to(device)
                lengths = batch['lengths']
                
                with autocast():
                    outputs = model(features, lengths)
                    batch_size, seq_len, num_classes = outputs.size()
                    outputs_flat = outputs.reshape(-1, num_classes)
                    labels_flat = labels.reshape(-1)
                    loss = criterion(outputs_flat, labels_flat)
                
                total_loss += loss.item()
                successful_batches += 1
                
                # Get predictions and probabilities
                probabilities = torch.softmax(outputs_flat, dim=1)
                predictions = torch.argmax(outputs_flat, dim=1)
                
                # Filter out padded tokens
                mask = (labels_flat != -100)
                if mask.sum() > 0:
                    valid_predictions = predictions[mask].cpu().numpy()
                    valid_labels = labels_flat[mask].cpu().numpy()
                    valid_probabilities = probabilities[mask].cpu().numpy()
                    
                    all_predictions.extend(valid_predictions)
                    all_labels.extend(valid_labels)
                    all_probabilities.extend(valid_probabilities)
                
                # Memory cleanup
                del features, labels, outputs, outputs_flat, labels_flat, probabilities, predictions
                if torch.cuda.is_available():
                    torch.cuda.empty_cache()
                    
                progress_bar.set_postfix({
                    'loss': f'{loss.item():.4f}',
                    'mem': f'{print_memory_usage(printing=False):.1f}GB'
                })
                
            except RuntimeError as e:
                if "out of memory" in str(e):
                    print(f"Skipping {phase.lower()} batch due to OOM")
                    if torch.cuda.is_available():
                        torch.cuda.empty_cache()
                    continue
                else:
                    raise e
    
    # Calculate metrics
    avg_loss = total_loss / successful_batches if successful_batches > 0 else float('inf')
    
    if len(all_predictions) > 0:
        # Basic metrics
        accuracy = accuracy_score(all_labels, all_predictions)
        precision, recall, f1, support = precision_recall_fscore_support(
            all_labels, all_predictions, average='weighted', zero_division=0
        )
        
        # Per-class metrics
        precision_per_class, recall_per_class, f1_per_class, support_per_class = precision_recall_fscore_support(
            all_labels, all_predictions, average=None, zero_division=0, labels=range(len(label_encoder.classes_))
        )
        
        # Classification report
        class_names = label_encoder.classes_
        report = classification_report(
            all_labels, all_predictions, 
            target_names=class_names, 
            zero_division=0,
            output_dict=True
        )
        
        metrics = {
            'loss': avg_loss,
            'accuracy': accuracy,
            'precision': precision,
            'recall': recall,
            'f1': f1,
            'precision_per_class': precision_per_class,
            'recall_per_class': recall_per_class,
            'f1_per_class': f1_per_class,
            'support_per_class': support_per_class,
            'classification_report': report,
            'predictions': all_predictions,
            'labels': all_labels,
            'probabilities': all_probabilities
        }
    else:
        metrics = {
            'loss': avg_loss,
            'accuracy': 0.0,
            'precision': 0.0,
            'recall': 0.0,
            'f1': 0.0
        }
    
    return metrics

def plot_confusion_matrix(y_true, y_pred, class_names, title="Confusion Matrix", save_path=None):
    """Plot and save confusion matrix."""
    cm = confusion_matrix(y_true, y_pred)
    
    plt.figure(figsize=(12, 10))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=class_names, yticklabels=class_names)
    plt.title(title)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"Confusion matrix saved to {save_path}")
    
    plt.show()

def plot_training_history(train_losses, val_losses, train_accuracies, val_accuracies, save_path=None):
    """Plot training history."""
    epochs = range(1, len(train_losses) + 1)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    
    # Plot losses
    ax1.plot(epochs, train_losses, 'b-', label='Training Loss')
    ax1.plot(epochs, val_losses, 'r-', label='Validation Loss')
    ax1.set_title('Training and Validation Loss')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')
    ax1.legend()
    ax1.grid(True)
    
    # Plot accuracies
    ax2.plot(epochs, train_accuracies, 'b-', label='Training Accuracy')
    ax2.plot(epochs, val_accuracies, 'r-', label='Validation Accuracy')
    ax2.set_title('Training and Validation Accuracy')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Accuracy')
    ax2.legend()
    ax2.grid(True)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"Training history saved to {save_path}")
    
    plt.show()

def save_detailed_results(metrics, label_encoder, output_dir, phase="test"):
    """Save detailed evaluation results."""
    results_path = os.path.join(output_dir, f"{phase}_results.json")
    
    # Convert numpy arrays to lists for JSON serialization
    json_metrics = {}
    for key, value in metrics.items():
        if isinstance(value, np.ndarray):
            json_metrics[key] = value.tolist()
        elif key == 'classification_report':
            json_metrics[key] = value
        elif key in ['predictions', 'labels', 'probabilities']:
            # Don't save raw predictions/labels/probabilities to JSON (too large)
            continue
        else:
            json_metrics[key] = value
    
    # Add class names
    json_metrics['class_names'] = label_encoder.classes_.tolist()
    
    with open(results_path, 'w') as f:
        json.dump(json_metrics, f, indent=2)
    
    print(f"Detailed results saved to {results_path}")
    
    # Save per-class results as CSV
    if 'precision_per_class' in metrics:
        per_class_df = pd.DataFrame({
            'class': label_encoder.classes_,
            'precision': metrics['precision_per_class'],
            'recall': metrics['recall_per_class'],
            'f1_score': metrics['f1_per_class'],
            'support': metrics['support_per_class']
        })
        
        per_class_path = os.path.join(output_dir, f"{phase}_per_class_metrics.csv")
        per_class_df.to_csv(per_class_path, index=False)
        print(f"Per-class metrics saved to {per_class_path}")

def train_with_lazy_loading(data_path, model_config, max_sequences=None):
    """Main training function with lazy loading, early stopping, and comprehensive evaluation."""
    
    # Create output directory
    output_dir = './models'
    os.makedirs(output_dir, exist_ok=True)
    
    # Create sequence metadata
    sequence_metadata, all_tags = create_sequence_metadata(data_path, max_sequences)
    
    # Create and fit label encoder
    label_encoder = LabelEncoder()
    label_encoder.fit(all_tags)
    num_classes = len(label_encoder.classes_)
    
    # Get input dimension from first sequence
    if data_path.endswith('.csv'):
        df_sample = pd.read_csv(data_path, nrows=1)
        df_sample['feature_vector'] = df_sample['feature_vector'].apply(ast.literal_eval).apply(np.array)
        input_dim = df_sample['feature_vector'].iloc[0].shape[0]
    elif data_path.endswith(('.hdf5', '.h5')):
        with h5py.File(data_path, 'r') as f:
            input_dim = f['feature_vector'][0].shape[0]
    
    print(f"Input dimension: {input_dim}")
    print(f"Number of classes: {num_classes}")
    print(f"Classes: {label_encoder.classes_}")
    
    # Split sequences: 70% train, 15% validation, 15% test
    train_metadata, temp_metadata = train_test_split(
        sequence_metadata, test_size=0.3, random_state=42
    )
    val_metadata, test_metadata = train_test_split(
        temp_metadata, test_size=0.5, random_state=42
    )
    
    print(f"Train sequences: {len(train_metadata)}")
    print(f"Validation sequences: {len(val_metadata)}")
    print(f"Test sequences: {len(test_metadata)}")
    
    # Create datasets with smaller cache sizes for memory efficiency
    cache_size = min(50, len(train_metadata) // 4)  # Adaptive cache size
    
    train_dataset = LazyFigmaDataset(
        data_path, train_metadata, label_encoder, cache_size=cache_size
    )
    val_dataset = LazyFigmaDataset(
        data_path, val_metadata, label_encoder, cache_size=cache_size//2
    )
    test_dataset = LazyFigmaDataset(
        data_path, test_metadata, label_encoder, cache_size=cache_size//2
    )
    
    # Create data loaders
    train_loader = DataLoader(
        train_dataset,
        batch_size=model_config['batch_size'],
        shuffle=True,
        collate_fn=lazy_collate_fn,
        num_workers=0,
        pin_memory=False
    )
    
    val_loader = DataLoader(
        val_dataset,
        batch_size=model_config['batch_size'],
        shuffle=False,
        collate_fn=lazy_collate_fn,
        num_workers=0,
        pin_memory=False
    )
    
    test_loader = DataLoader(
        test_dataset,
        batch_size=model_config['batch_size'],
        shuffle=False,
        collate_fn=lazy_collate_fn,
        num_workers=0,
        pin_memory=False
    )
    
    # Create model
    model = OptimizedFigmaBLSTM(
        input_dim=input_dim,
        hidden_dim=model_config['hidden_dim'],
        output_dim=num_classes,
        num_layers=model_config['num_layers'],
        dropout=model_config['dropout']
    ).to(device)
    
    # Training setup
    criterion = nn.CrossEntropyLoss(ignore_index=-100)
    optimizer = optim.Adam(model.parameters(), lr=model_config['learning_rate'])
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=3)
    
    # Early stopping
    early_stopping = EarlyStopping(
        patience=model_config['early_stopping_patience'],
        min_delta=0.0001,
        restore_best_weights=True
    )
    
    # Training history
    train_losses = []
    val_losses = []
    train_accuracies = []
    val_accuracies = []
    
    print("Starting training with lazy loading...")
    print("=" * 50)
    
    start_time = time.time()
    
    for epoch in range(model_config['epochs']):
        epoch_start_time = time.time()
        print(f"Epoch {epoch+1}/{model_config['epochs']}")
        print_memory_usage()
        
        # Training phase
        model.train()
        train_loss = 0.0
        successful_batches = 0
        
        train_progress = tqdm(train_loader, desc="Training")
        for batch_idx, batch in enumerate(train_progress):
            loss = safe_training_step(model, batch, criterion, optimizer, device)
            
            if loss is not None:
                train_loss += loss
                successful_batches += 1
                train_progress.set_postfix({
                    'loss': f'{loss:.4f}',
                    'mem': f'{print_memory_usage(printing=False):.1f}GB'
                })
            
            # Periodic cleanup
            if batch_idx % 10 == 0:
                gc.collect()
        
        avg_train_loss = train_loss / successful_batches if successful_batches > 0 else float('inf')
        
        # Validation phase
        val_metrics = evaluate_model(model, val_loader, criterion, device, label_encoder, "Validation")
        
        # Calculate training accuracy on a subset for efficiency
        train_metrics = evaluate_model(model, train_loader, criterion, device, label_encoder, "Training")
        
        # Update history
        train_losses.append(avg_train_loss)
        val_losses.append(val_metrics['loss'])
        train_accuracies.append(train_metrics['accuracy'])
        val_accuracies.append(val_metrics['accuracy'])
        
        # Print epoch results
        epoch_time = time.time() - epoch_start_time
        print(f"Epoch {epoch+1} completed in {epoch_time:.2f}s")
        print(f"Train Loss: {avg_train_loss:.4f}, Train Acc: {train_metrics['accuracy']:.4f}")
        print(f"Val Loss: {val_metrics['loss']:.4f}, Val Acc: {val_metrics['accuracy']:.4f}")
        print(f"Val Precision: {val_metrics['precision']:.4f}, Val Recall: {val_metrics['recall']:.4f}, Val F1: {val_metrics['f1']:.4f}")
        
        # Save best model
        if val_metrics['loss'] < float('inf'):
            model_path = os.path.join(output_dir, "figma_blstm_lazy_model.pt")
            
            torch.save({
                'epoch': epoch,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'val_loss': val_metrics['loss'],
                'val_accuracy': val_metrics['accuracy'],
                'label_encoder': label_encoder,
                'model_config': model_config,
                'input_dim': input_dim,
                'num_classes': num_classes,
                'sequence_metadata': sequence_metadata,
                'train_losses': train_losses,
                'val_losses': val_losses,
                'train_accuracies': train_accuracies,
                'val_accuracies': val_accuracies
            }, model_path)
            print(f"Model saved to {model_path}")
        
        # Learning rate scheduling
        scheduler.step(val_metrics['loss'])
        
        # Early stopping check
        if early_stopping(val_metrics['loss'], model):
            print(f"Early stopping triggered at epoch {epoch+1}")
            break
        
        print("-" * 50)
        
        # Force garbage collection
        gc.collect()
        if torch.cuda.is_available():
            torch.cuda.empty_cache()
    
    total_training_time = time.time() - start_time
    print(f"Training completed in {total_training_time:.2f} seconds")
    
    # Plot training history
    plot_training_history(
        train_losses, val_losses, train_accuracies, val_accuracies,
        save_path=os.path.join(output_dir, "training_history.png")
    )
    
    # Final evaluation on test set
    print("\n" + "="*50)
    print("FINAL EVALUATION ON TEST SET")
    print("="*50)
    
    test_metrics = evaluate_model(model, test_loader, criterion, device, label_encoder, "Test")
    
    print(f"Test Results:")
    print(f"  Loss: {test_metrics['loss']:.4f}")
    print(f"  Accuracy: {test_metrics['accuracy']:.4f} ({test_metrics['accuracy']*100:.2f}%)")
    print(f"  Precision: {test_metrics['precision']:.4f}")
    print(f"  Recall: {test_metrics['recall']:.4f}")
    print(f"  F1-Score: {test_metrics['f1']:.4f}")
    
    # Print per-class results for important classes
    if 'precision_per_class' in test_metrics:
        print(f"\nPer-class Test Results:")
        for i, class_name in enumerate(label_encoder.classes_):
            if test_metrics['support_per_class'][i] > 0:  # Only show classes with samples
                print(f"  {class_name}:")
                print(f"    Precision: {test_metrics['precision_per_class'][i]:.4f}")
                print(f"    Recall: {test_metrics['recall_per_class'][i]:.4f}")
                print(f"    F1-Score: {test_metrics['f1_per_class'][i]:.4f}")
                print(f"    Support: {test_metrics['support_per_class'][i]}")
    
    # Save detailed results
    save_detailed_results(test_metrics, label_encoder, output_dir, "test")
    
    # Plot confusion matrix
    if len(test_metrics['predictions']) > 0:
        plot_confusion_matrix(
            test_metrics['labels'], 
            test_metrics['predictions'],
            label_encoder.classes_,
            title="Test Set Confusion Matrix",
            save_path=os.path.join(output_dir, "test_confusion_matrix.png")
        )
    
    # Save final model with all metrics
    final_model_path = os.path.join(output_dir, "figma_blstm_final_model.pt")
    torch.save({
        'model_state_dict': model.state_dict(),
        'label_encoder': label_encoder,
        'model_config': model_config,
        'input_dim': input_dim,
        'num_classes': num_classes,
        'sequence_metadata': sequence_metadata,
        'train_losses': train_losses,
        'val_losses': val_losses,
        'train_accuracies': train_accuracies,
        'val_accuracies': val_accuracies,
        'test_metrics': test_metrics,
        'training_time': total_training_time
    }, final_model_path)
    print(f"Final model with all metrics saved to {final_model_path}")
    
    # Print training summary
    print(f"\n" + "="*50)
    print("TRAINING SUMMARY")
    print("="*50)
    print(f"Total Training Time: {total_training_time:.2f} seconds")
    print(f"Total Epochs: {len(train_losses)}")
    print(f"Best Validation Loss: {min(val_losses):.4f}")
    print(f"Best Validation Accuracy: {max(val_accuracies):.4f}")
    print(f"Final Test Accuracy: {test_metrics['accuracy']:.4f}")
    print(f"Final Test F1-Score: {test_metrics['f1']:.4f}")
    
    # Cleanup datasets
    train_dataset._close_file()
    val_dataset._close_file()
    test_dataset._close_file()
    
    return model, label_encoder, test_metrics

def load_and_evaluate_model(model_path, data_path, test_sequences=None):
    """Load a saved model and evaluate it on test data."""
    print("Loading saved model...")
    
    checkpoint = torch.load(model_path, map_location=device)
    
    # Extract model configuration
    model_config = checkpoint['model_config']
    input_dim = checkpoint['input_dim']
    num_classes = checkpoint['num_classes']
    label_encoder = checkpoint['label_encoder']
    
    # Create model
    model = OptimizedFigmaBLSTM(
        input_dim=input_dim,
        hidden_dim=model_config['hidden_dim'],
        output_dim=num_classes,
        num_layers=model_config['num_layers'],
        dropout=model_config['dropout']
    ).to(device)
    
    # Load model weights
    model.load_state_dict(checkpoint['model_state_dict'])
    
    print("Model loaded successfully!")
    
    # If test sequences are provided, evaluate on them
    if test_sequences is not None:
        test_dataset = LazyFigmaDataset(data_path, test_sequences, label_encoder, cache_size=25)
        test_loader = DataLoader(
            test_dataset,
            batch_size=model_config['batch_size'],
            shuffle=False,
            collate_fn=lazy_collate_fn,
            num_workers=0,
            pin_memory=False
        )
        
        criterion = nn.CrossEntropyLoss(ignore_index=-100)
        test_metrics = evaluate_model(model, test_loader, criterion, device, label_encoder, "Test")
        
        test_dataset._close_file()
        return model, label_encoder, test_metrics
    
    return model, label_encoder, None

# Example usage
if __name__ == "__main__":
    # Enhanced configuration with more epochs and early stopping
    model_config = {
        'hidden_dim': 128,
        'num_layers': 2,  # Increased layers
        'dropout': 0.3,
        'learning_rate': 0.001,
        'batch_size': 4,  # Smaller batch size for memory efficiency
        'epochs': 50,  # Increased epochs
        'early_stopping_patience': 10  # Early stopping patience
    }
    
    DATA_PATH = "figma_dataset_custom.h5"  # Update this path
    MAX_SEQUENCES = 2000  # Adjust based on your needs
    
    print("Starting enhanced training with comprehensive evaluation...")
    print(f"Configuration: {model_config}")
    
    # Train model with lazy loading and comprehensive evaluation
    model, label_encoder, test_metrics = train_with_lazy_loading(
        DATA_PATH, model_config, max_sequences=MAX_SEQUENCES
    )
    
    print("\n" + "="*50)
    print("TRAINING COMPLETED SUCCESSFULLY!")
    print("="*50)
    print("Files saved:")
    print("  - figma_blstm_final_model.pt (final model)")
    print("  - training_history.png (training curves)")
    print("  - test_confusion_matrix.png (confusion matrix)")
    print("  - test_results.json (detailed metrics)")
    print("  - test_per_class_metrics.csv (per-class performance)")
    
    final_memory = print_memory_usage()
    print(f"Final memory usage: {final_memory:.2f} GB")
```
